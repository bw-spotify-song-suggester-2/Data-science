```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KDTree
import os
import pickle
```

```python
data_path = 'track_master_df.csv'
train_df = pd.read_csv(data_path)
test_df = pd.read_csv('test_case.csv')

drop_cols = ['artist', 'album', 'track', 'track_id']

X_train = train_df.drop(columns = drop_cols)
X_test = test_df.drop(columns = drop_cols)
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
tree = KDTree(X_train, leaf_size=1)
s = pickle.dumps(tree)
tree_copy = pickle.loads(s)       
dist, ind = tree_copy.query(X_test, k=10)   
```

```python
with open('kdtree_model_1.pkl', 'wb') as f:
    pickle.dump(tree, f)
```
